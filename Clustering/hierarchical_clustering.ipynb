```python
import pandas as pd
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:, [3,4]].values
```

### Use dendogram to find Optimal number of clusters

```python
from scipy.cluster import hierarchy as hrc
from matplotlib import pyplot as plt 

# Plot the dendrogram
plt.figure(figsize=(10, 7))

dendogram = hrc.dendrogram(hrc.linkage(x, method='ward'))

plt.title('Dendrogram')
plt.xlabel('Customer ID (Observation point)')
plt.ylabel('Euclidean Distance')

threshold_value = 150  # Adjust this value based on your dendrogram
plt.axhline(y=threshold_value, color='r', linestyle='--')
plt.show()
```

### Training hierarchical cluster on database

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean',linkage='ward')
Y_pred = cluster.fit_predict(x)
print(Y_pred)
```

```python
plt.scatter(x[Y_pred==0, 0], x[Y_pred==0,1], label="cluster-0", c='red')
plt.scatter(x[Y_pred==1, 0], x[Y_pred==1,1], label="cluster-1", c='green')
plt.scatter(x[Y_pred==2, 0], x[Y_pred==2,1], label="cluster-2", c='blue')
plt.scatter(x[Y_pred==3, 0], x[Y_pred==3,1], label="cluster-3", c='cyan')
plt.scatter(x[Y_pred==4, 0], x[Y_pred==4,1], label="cluster-4", c='magenta')
plt.title('Clusters of customer')
plt.xlabel('Annual income')
plt.ylabel('Spending Score 1-100')
plt.legend()
plt.show()
```
